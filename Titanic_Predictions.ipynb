# PREDICTING SURVIVAL RATE IN TITANIC TRAGEDY

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('titanic.csv')
df.head(5)
```

```python
df=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis=1)
```

```python
df.head(5)
```

```python
df.columns[df.isna().any()]
```

```python
x=df['Age'].mean()
y=int(x)
df['Age']=df['Age'].fillna(y)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Sex']=le.fit_transform(df['Sex'])
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
x1,x2,y1,y2=train_test_split(df[['Pclass','Sex','Age']],df[['Survived']],train_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(x1,y1)
```

```python
pd.DataFrame(reg.predict(x2)).head(5)
```

```python
reg.score(x2,y2)
```
